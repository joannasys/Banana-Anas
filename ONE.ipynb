```python
import glob
import librosa
from scipy.io import wavfile
import re
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from PIL import Image
import os
```

```python
MMS = MinMaxScaler()

#spectrogram = []
y = []

for filename in glob.iglob('./train/training/*'):
    for i,filename_wav in enumerate(glob.iglob(filename+'/*')):
      
        sample_rate, samples = wavfile.read(filename_wav)
        S = librosa.feature.melspectrogram(samples, sr=sample_rate, n_mels=128)
        log_S = librosa.power_to_db(S, ref=np.max)
        
        if log_S.shape == (128,32):         # only train data that have 32 columns
            
            log_S_S = MMS.fit_transform(log_S)   # Normalisation of values
            X = np.stack((log_S_S,)*3,axis=2)     
            rescaled = (255.0 /X.max() * (X - X.min())).astype(np.uint8)
            
            #spectrogram.append(log_S_S)   # list of spectrogram array
            
            im = Image.fromarray(rescaled)
            
            label = re.search('[^/]+$',filename).group()   # class
            name = re.search('([^/]*).wav$',filename_wav).group(1)
            
            labelpath = './train/image/'+str(label)
            if not os.path.isdir(labelpath):               # if folder doesn't exist, make new folder in class label name
                os.makedirs(labelpath)

            im.save('./train/image/'+str(label)+'/'+str(name)+'.jpeg')
            
            y.append(label)     # list of labels
            
        if i%100 == 0:
            print('saved image {}'.format(i+100))
```

```python
import pickle

with open('y.pickle', 'wb') as f:
    pickle.dump(y, f)
```
